```python
import pandas as pd
import numpy as np
start_year = 2015
```

```python
#Creating Language DataFrame
language_file2015 = 'Resources/2015/data/language_2015.csv'
language_file2016 = 'Resources/2016/data/language_2016.csv'
language_file2017 = 'Resources/2017/data/language_2017.csv'
language_file2018 = 'Resources/2018/data/language_2018.csv'

language_2015 = pd.read_csv(language_file2015)
language_2016 = pd.read_csv(language_file2016)
language_2017 = pd.read_csv(language_file2017)
language_2018 = pd.read_csv(language_file2018)

language_2015 = pd.DataFrame(language_2015)
language_2016 = pd.DataFrame(language_2016)
language_2017 = pd.DataFrame(language_2017)
language_2018 = pd.DataFrame(language_2018)

language_datas = [language_2015, language_2016, language_2017, language_2018]

spanish_household_data = language_2015.copy()
spanish_household_data = spanish_household_data.loc[:,["GEO_ID","NAME"]]
spanish_household_data = spanish_household_data.rename(columns={"NAME":"State"})

spanish_only_data = language_2015.copy()
spanish_only_data = spanish_only_data.loc[:,["GEO_ID","NAME"]]
spanish_only_data = spanish_only_data.rename(columns={"NAME":"State"})

language_data_labels =["NAME","S1602_C02_002E","S1602_C04_002E",]
year = start_year
for data in language_datas: 
    data = data[language_data_labels]
    for x in range(len(language_data_labels)):
        if x > 0:
            try:
                data.iloc[:,x] = pd.to_numeric(data.iloc[:,x])
            except ValueError:
                None
    data = data.rename(columns={language_data_labels[0]:"State",
                                language_data_labels[1]:"{}".format(year),
                                language_data_labels[2]:"{}".format(year)})
    spanish_household_data = pd.merge(spanish_household_data, data.iloc[:,[0,1]], on="State")
    spanish_only_data = pd.merge(spanish_only_data, data.iloc[:,[0,2]], on="State")
    year += 1
spanish_household_data = spanish_household_data.iloc[1:,1:]
spanish_household_data.to_csv("output/pop_spanish_household.csv", index=False, header=True)
spanish_only_data = spanish_only_data.iloc[1:,1:]
spanish_only_data.to_csv("output/pop_spanish_only.csv", index=False, header=True)
spanish_household_data
```

```python
#Creating Proverty DataFrame
poverty_file2015 = 'Resources/2015/data/poverty_2015.csv'
poverty_file2016 = 'Resources/2016/data/poverty_2016.csv'
poverty_file2017 = 'Resources/2017/data/poverty_2017.csv'
poverty_file2018 = 'Resources/2018/data/poverty_2018.csv'

poverty_2015 = pd.read_csv(poverty_file2015)
poverty_2016 = pd.read_csv(poverty_file2016)
poverty_2017 = pd.read_csv(poverty_file2017)
poverty_2018 = pd.read_csv(poverty_file2018)

poverty_2015 = pd.DataFrame(poverty_2015)
poverty_2016 = pd.DataFrame(poverty_2016)
poverty_2017 = pd.DataFrame(poverty_2017)
poverty_2018 = pd.DataFrame(poverty_2018)

poverty_datas = [poverty_2015, poverty_2016, poverty_2017, poverty_2018]
poverty_data_all = poverty_2015.copy()
poverty_data_all = poverty_data_all.loc[:,["GEO_ID","NAME"]]
poverty_data_all = poverty_data_all.rename(columns={"NAME":"State"})
poverty_data_labels = ["NAME","S1701_C03_001E"]

year = start_year
for data in poverty_datas: 
    data = data[poverty_data_labels]
    for x in range(len(poverty_data_labels)):
        if x > 0:
            try:
                data.iloc[:,x] = pd.to_numeric(data.iloc[:,x])
            except ValueError:
                None
    data = data.rename(columns={poverty_data_labels[0]:"State",
                                poverty_data_labels[1]:"{}".format(year)})
    poverty_data_all = pd.merge(poverty_data_all, data, on="State")
    year += 1
poverty_data_all_final = poverty_data_all.iloc[1:,1:]
poverty_data_all_final.to_csv("output/pop_poverty.csv", index=False, header=True)
poverty_data_all_final
```

```python
#Age Demographic Dataset
age_file2015 = 'Resources/2015/data/age_2015.csv'
age_file2016 = 'Resources/2016/data/age_2016.csv'
age_file2017 = 'Resources/2017/data/age_2017.csv'
age_file2018 = 'Resources/2018/data/age_2018.csv'

age_2015 = pd.read_csv(age_file2015)
age_2016 = pd.read_csv(age_file2016)
age_2017 = pd.read_csv(age_file2017)
age_2018 = pd.read_csv(age_file2018)

age_2015 = pd.DataFrame(age_2015)
age_2016 = pd.DataFrame(age_2016)
age_2017 = pd.DataFrame(age_2017)
age_2018 = pd.DataFrame(age_2018)

age_datas = [age_2015, age_2016, age_2017, age_2018]
age_under18_data = age_2015.copy()
age_under18_data = age_under18_data.loc[:,["GEO_ID","NAME"]]
age_under18_data = age_under18_data.rename(columns={"NAME":"State"})
age_18to64_data = age_2015.copy()
age_18to64_data = age_18to64_data.loc[:,["GEO_ID","NAME"]]
age_18to64_data = age_18to64_data.rename(columns={"NAME":"State"})
age_65andup_data = age_2015.copy()
age_65andup_data = age_65andup_data.loc[:,["GEO_ID","NAME"]]
age_65andup_data = age_65andup_data.rename(columns={"NAME":"State"})


year = start_year
for data in age_datas: 
    if year > 2016:
        data = data.iloc[1:,1:]
        age_data_used = ["S0101_C02_022E","S0101_C02_026E","S0101_C02_030E"]
        age_data_labels = ["NAME","S0101_C02_022E","18-64","S0101_C02_030E"]
        for column in age_data_used:
            try:
                data[column] = pd.to_numeric(data[column])
            except ValueError:
                None
        data["18-64"]=data["S0101_C02_026E"]-data["S0101_C02_030E"]
    else:
        data = data.iloc[1:,1:]
        data = data.reset_index(drop=True)
        age_data_used = ["S0101_C01_001E","S0101_C01_002E","S0101_C01_020E","S0101_C01_021E","S0101_C01_025E","S0101_C01_028E"]
        for column in age_data_used:
            try:
                data[column] = pd.to_numeric(data[column])
            except ValueError:
                None
        age_data_labels = ["NAME","under 18","18-64","65 and up"]
        data["under 18"] = (data["S0101_C01_002E"]+data["S0101_C01_020E"]+data["S0101_C01_021E"])
        data["18-64"] = (data["S0101_C01_025E"]-data["S0101_C01_028E"])
        data["65 and up"]  = data["S0101_C01_028E"]
    data = data[age_data_labels]
    data = data.rename(columns={age_data_labels[0]:"State",
                                age_data_labels[1]:"{}".format(year),
                                age_data_labels[2]:"{}".format(year),
                                age_data_labels[3]:"{}".format(year)})
    age_under18_data = pd.merge(age_under18_data, data.iloc[:,[0,1]], on="State")
    age_18to64_data = pd.merge(age_18to64_data, data.iloc[:,[0,2]], on="State")
    age_65andup_data = pd.merge(age_65andup_data, data.iloc[:,[0,3]], on="State")
    year += 1


age_under18_data = age_under18_data.iloc[:,1:]
age_under18_data = age_under18_data.reset_index(drop=True)
age_18to64_data = age_18to64_data.iloc[:,1:]
age_18to64_data = age_18to64_data.reset_index(drop=True)
age_65andup_data = age_65andup_data.iloc[:,1:]
age_65andup_data = age_65andup_data.reset_index(drop=True)

age_under18_data.to_csv("output/pop_age_under18.csv", index=False, header=True)
age_18to64_data.to_csv("output/pop_age_18to64_data.csv", index=False, header=True)
age_65andup_data.to_csv("output/pop_age_65andup.csv", index=False, header=True)
```
