```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
%matplotlib notebook
year_used = "2016"
pop_veteran = pd.read_csv(f"output/pop_veteran_{year_used}.csv")
pop_education = pd.read_csv(f"output/pop_education_{year_used}.csv")
pop_suicide = pd.read_csv(f"output/pop_suicide_{year_used}.csv")
```

```python
#Loop through list of states in order of education trends. Create a list of colors. Blue for democrat, red for republican.
party = []
index_count = 0
for rows in pop_education.iterrows(): 
    if pop_education.loc[index_count, 'Party']== 'D':
        party.append('royalblue')
    elif pop_education.loc[index_count, 'Party']== 'R':
        party.append('firebrick')
    else:
        party.append('darkmagenta') 
    index_count = index_count + 1
    
x_ticks = [i for i, _ in enumerate(pop_education['State'])]

plt.bar(x_ticks, pop_education[year_used], .6, color=party)

plt.ylabel("% of Bachelor Degree Holders")
plt.title("% of Bachelor Degree Holders in 2016 by State and Party Affiliation", wrap = True)

plt.xticks(x_ticks, pop_education['State'], rotation = 90)
plt.tight_layout()
plt.style.use('seaborn')

#Calculating Means of Each Party
pop_education_party = pop_education.groupby(["Party"]).mean()
means = {}
for index in pop_education_party.index.tolist():
    means[index]="{:.2f}".format(pop_education_party.loc[index,year_used])

#Create legend for party affiliation
red_patch = mpatches.Patch(color='firebrick', label=f'Republican (mean={means["R"]})')
blue_patch = mpatches.Patch(color='royalblue', label=f'Democrat (mean={means["D"]})')

plt.legend("best", handles=[red_patch, blue_patch])

plt.savefig(f"Images/education_population{year_used}.png")
plt.show()
```

```python
#Loop through list of states in order of suicide trends. Create a list of colors. Blue for democrat, red for republican.
party = []
index_count = 0
for rows in pop_suicide.iterrows(): 
    if  pop_suicide.loc[index_count, 'Party']== 'D':
        party.append('royalblue')
    elif pop_suicide.loc[index_count, 'Party']== 'R':
        party.append('firebrick')
    else:
        party.append('darkmagenta') 
    index_count = index_count + 1
    
x_ticks = [i for i, _ in enumerate(pop_suicide['State'])]

plt.bar(x_ticks, pop_suicide[year_used], .6, color=party)

plt.ylabel("Suicide Rate per 100,000")
plt.title("Suicide Rate in 2016 by State and Party Affiliation", wrap = True)

plt.xticks(x_ticks, pop_suicide['State'], rotation = 90)
plt.tight_layout()
plt.style.use('seaborn')

#Calculating Means of Each Party
pop_suicide_party = pop_suicide.groupby(["Party"]).mean()
means = {}
for index in pop_suicide_party.index.tolist():
    means[index]="{:.2f}".format(pop_suicide_party.loc[index,year_used])

#Create legend for party affiliation
red_patch = mpatches.Patch(color='firebrick', label=f'Republican (mean={means["R"]})')
blue_patch = mpatches.Patch(color='royalblue', label=f'Democrat (mean={means["D"]})')


plt.legend("best", handles=[red_patch, blue_patch])

plt.savefig("Images/suicide_population.png")
plt.show()
```

```python
#Loop through list of states in order of veteran trends. Create a list of colors. Blue for democrat, red for republican.
party = []
index_count = 0
for rows in pop_veteran.iterrows(): 
    if  pop_veteran.loc[index_count, 'Party']== 'D':
        party.append('royalblue')
    elif pop_veteran.loc[index_count, 'Party']== 'R':
        party.append('firebrick')
    else:
        party.append('darkmagenta') 
    index_count = index_count + 1
    
x_ticks = [i for i, _ in enumerate(pop_veteran['State'])]

plt.bar(x_ticks, pop_veteran[year_used], .6, color=party)

plt.ylabel("Veteran Population by %")
plt.title("Veteran Percentage in 2016 by State and Party Affiliation", wrap = True)

plt.xticks(x_ticks, pop_veteran['State'], rotation = 90)
plt.tight_layout()
plt.style.use('seaborn')

#Calculating Means of Each Party
pop_veteran_party = pop_veteran.groupby(["Party"]).mean()
means = {}
for index in pop_veteran_party.index.tolist():
    means[index]="{:.2f}".format(pop_veteran_party.loc[index,year_used])

#Create legend for party affiliation
red_patch = mpatches.Patch(color='firebrick', label=f'Republican (mean={means["R"]})')
blue_patch = mpatches.Patch(color='royalblue', label=f'Democrat (mean={means["D"]})')
purple_patch = mpatches.Patch(color='darkmagenta', label=f'Independent (mean={means["I"]})')

plt.legend("best", handles=[red_patch, blue_patch, purple_patch])

plt.savefig("Images/veteran_population.png")
plt.show()
```
